```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
temp = [25, 35, 45, 55, 65, 75]
Vf_temp = [0.221, 0.188, 0.156, 0.125, 0.096, 0.071]
Ir_temp = [160.2, 161.6, 165, 173, 190.8, 228.6]
```

```python
plt.plot(temp, Vf_temp)
plt.title("Temperature variation of Vf")
plt.show()
plt.plot(temp, Ir_temp)
plt.title("Temperature variation of Ir")
plt.show()
```

As temperature increases, the current in the diode also increases, because of which the forward voltage reduces. Both the forward and the reverse currents increase with temperature, therefore we see an exponential rise in reverse current

```python
ill = [0.5, 1, 1.5, 2, 2.5, 3]
Vf_ill = [0.209, 0.221, 0.229, 0.237, 0.243, 0.248]
Ir_ill = [110.4, 160.2, 209.9, 259.8, 309.6, 359.4]
```

```python
plt.plot(ill, Vf_ill)
plt.title("Illumination variation of Vf")
plt.show()
plt.plot(ill, Ir_ill)
plt.title("Illumination variation of Ir")
plt.show()
```

As the Illumination level increases, the I-V plot shifts downward, because of which the forward voltage increases as we need to apply higher voltage for the same current but due to the downward shift, reverse current increases. Both increases are linear in nature
